```python
file_path='path_to_your_excel_file.xlsx'
data=pd.read_excel(file_path)
```

```python
features=data[['feature_1', 'feature_2', 'feature_3', 'feature_4', 'feature_5']]
target=data['target']
```

```python
X_train, X_test, y_train, y_test =train_test_split(features, target, test_size=0.9, random_state=12)
```

```python
import sklearn.preprocessing
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
import keras
import tensorflow as tf
model=tf.keras.Sequential([tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
                           tf.keras.layers.Dense(32, activation='relu'),
                           tf.keras.layers.Dense(1, activation='sigmoid')])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history=model.fit(X_train, y_train, epochs=100, validation_split=0.2)
```

```python
loss, accuracy=model.evaluate(X_test, y_test)
print(f'Тестовая точность:{accuracy:.4f}')
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='Тренировочная потеря')
plt.plot(history.history['val_loss'], label='Потеря валидации')
plt.title('Потеря во время обучения')
plt.xlabel('Эпоха')
plt.ylabel('Потеря')
plt.legend()
plt.show()
```
